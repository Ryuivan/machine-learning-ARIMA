### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Preprocessing

#### Dataset 2019

```python
# Import 2019 dataset
data_2019 = pd.read_csv("Jumlah Perjalanan Wisatawan Nusantara Menurut Provinsi Tujuan, 2019.csv", header=None)
print("Total instance and features:", data_2019.shape)
print("List of features:", data_2019.columns)
print("Print the first 5 instance:\n", data_2019.head())
```

#### Dataset 2020

```python
# Import 2020 dataset
data_2020 = pd.read_csv("Jumlah Perjalanan Wisatawan Nusantara Menurut Provinsi Tujuan, 2020.csv", header=None)
print("Total instance and features:", data_2020.shape)
print("List of features:", data_2020.columns)
print("Print the first 5 instance:\n", data_2020.head())
```

#### Dataset 2021

```python
# Import 2021 dataset
data_2021 = pd.read_csv("Jumlah Perjalanan Wisatawan Nusantara Menurut Provinsi Tujuan, 2021.csv", header=None)
print("Total instance and features:", data_2021.shape)
print("List of features:", data_2021.columns)
print("Print the first 5 instance:\n", data_2021.head())
```

#### Dataset 2022

```python
# Import 2022 dataset
data_2022 = pd.read_csv("Jumlah Perjalanan Wisatawan Nusantara Menurut Provinsi Tujuan, 2022.csv", header=None)
print("Total instance and features:", data_2022.shape)
print("List of features:", data_2022.columns)
print("Print the first 5 instance:\n", data_2022.head())
```

#### Dataset 2023

```python
# Import 2023 dataset
data_2023 = pd.read_csv("Jumlah Perjalanan Wisatawan Nusantara Menurut Provinsi Tujuan, 2023.csv", header=None)
print("Total instance and features:", data_2023.shape)
print("List of features:", data_2023.columns)
print("Print the first 5 instance:\n", data_2023.head())
```

#### Merging the 2019-2023 dataset

##### Drop unnecessary features

```python
# Drop 'Tahunan' features for 2019 dataset
data_2019 = data_2019.drop(columns=[13]) 

# Drop 'Provinsi' and 'Tahunan' features for 2020-2023 dataset
data_2020 = data_2020.drop(columns=[0, 13])  
data_2021 = data_2021.drop(columns=[0, 13])
data_2022 = data_2022.drop(columns=[0, 13])
data_2023 = data_2023.drop(columns=[0, 13])  

# Print total of features after drop
print("Total features after drop (2019):", data_2019.shape[1])
print("Total features after drop (2020):", data_2020.shape[1])
print("Total features after drop (2021):", data_2021.shape[1])
print("Total features after drop (2022):", data_2022.shape[1])
print("Total features after drop (2023):", data_2023.shape[1])
```

##### Combine/concat the datasets (2019-2023)

```python
# Combine the datasets using concat along features (axis=1)
data = pd.concat([data_2019, data_2020, data_2021, data_2022, data_2023], axis=1)

# Drop instance 0, 1, and 2 which contain the titles and descriptions of the dataset
data = data.drop([0, 1, 2])

# Reset index
data.reset_index(drop=True, inplace=True)

# Print the features of the combined dataset
print("List of features:", data.columns)
print("Total features:", data.shape[1])
```

##### Create new feature names after concatenating datasets

```python
# Create new feature names for the concatenated dataset
months = ['january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december']
years = ['2019', '2020', '2021', '2022', '2023']

# Rename features to ['Province'] + ['{month}'] * 12 + ['Annual']
new_features = ['province'] + [f'{month}_{year}' for year in years for month in months]

# Assign new features names
data.columns = new_features 

data.columns
```

##### Handling missing values

```python
# Find missing values ('-') in the dataset
missing_value = data[data.eq('-').any(axis=1)]

# Print the province that contains missing values
print("Province with missing values:", missing_value['province'].unique())

# Replace missing values ('-') with NaN
data.replace('-', np.nan, inplace=True)

# Drop NaN values
data.dropna(inplace=True)

# Print total of instances after drop
print("\nTotal instance after drop:", data.shape[0])

# Print list of province feature after drop
print(data['province'].unique())
```

#### Drop unnecessary instance

```python
# Drop unnecessary instance (INDONESIA)
data.drop(data[data['province'] == 'INDONESIA'].index, inplace=True)

# Display the final dataset shape and features
print("Total instance and features:", data.shape)
print("\nList of features:", data.columns)
print("\nList of instance (province):", data['province'].unique())
```

### Feature engineering

#### Creating growth_rate feature for each province per year

```python
# Converting numeric features (starting from index 1) to numeric type
data.iloc[:, 1:] = data.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Calculating the total visits per year for each province
total_2019 = data.iloc[:, 1:13].sum(axis=1)  # Total for 2019
total_2020 = data.iloc[:, 13:25].sum(axis=1)  # Total for 2020
total_2021 = data.iloc[:, 25:37].sum(axis=1)  # Total for 2021
total_2022 = data.iloc[:, 37:49].sum(axis=1)  # Total for 2022
total_2023 = data.iloc[:, 49:61].sum(axis=1)  # Total for 2023

# Calculating the Growth Rate from year to year
data['growth_rate_2020'] = (
    (total_2020 - total_2019) / total_2019.replace(0, pd.NA) * 100
)
data['growth_rate_2021'] = (
    (total_2021 - total_2020) / total_2020.replace(0, pd.NA) * 100
)
data['growth_rate_2022'] = (
    (total_2022 - total_2021) / total_2021.replace(0, pd.NA) * 100
)
data['growth_rate_2023'] = (
    (total_2023 - total_2022) / total_2022.replace(0, pd.NA) * 100
)

# Extracting growth rate data from 2020 to 2023
growth_rates = data[['growth_rate_2020', 'growth_rate_2021', 'growth_rate_2022', 'growth_rate_2023']]

# Concatenate province names and growth rates
result = pd.concat([data['province'], growth_rates], axis=1)

# Print the growth rate data for each province
print(result.head())
```

##### Creating and importing new data

```python
# Creating a CSV file from the preprocessed data
data.to_csv('Jumlah Perjalanan Wisatawan Nusantara Menurut Provinsi Tujuan, 2019-2023.csv', index=False)

# Importing data using the preprocessed file
data = pd.read_csv('Jumlah Perjalanan Wisatawan Nusantara Menurut Provinsi Tujuan, 2019-2023.csv')

# Print data
print(data.head())
```

#### Scaling with RobustScaler

```python
from sklearn.preprocessing import RobustScaler

# Converting numeric features (ignoring the 'province' and 'growth_rates' feature) to numeric types
data_for_scaling = data.iloc[:, 1:-4].apply(pd.to_numeric, errors='coerce') 

# Initialize RobustScaler
robust = RobustScaler()

# Apply the scaler to the data
data_scaled = robust.fit_transform(data_for_scaling)

# Convert the result into a DataFrame with the same feature names for the first 12 features
data_scaled = pd.DataFrame(data_scaled, columns=data.columns[1:-4])

# Display the data with the specified features scaled
print(data_scaled.head())
```

#### Creating heatmaps

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Defining the years to analyze
years = ['2019', '2020', '2021', '2022', '2023']
months = ['january', 'february', 'march', 'april', 'may', 'june', 
          'july', 'august', 'september', 'october', 'november', 'december']

# Creating a heatmap for each year
for year in years:
    # Calculating total visits for the specific year
    monthly_visits = data[[f'{month}_{year}' for month in months]].sum()
    
    # Reshaping the data into a DataFrame format for the heatmap
    heatmap_data = monthly_visits.values.reshape(-1, 1)

    # Creating the heatmap
    plt.figure(figsize=(5, 6))
    sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.2f', cbar_kws={'label': 'Total Visits'},
                yticklabels=months)
    plt.title(f'Heatmap of Monthly Visits - {year}')
    plt.xlabel('Total Visits')
    plt.ylabel('Month')
    plt.show()
```

##### Mean of domestic tourist visits per month
